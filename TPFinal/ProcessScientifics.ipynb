```python
import json
```

```python
papers = []
for year in range(2000, 2010):
    with open(f"data/papers_{year}.json", "r") as f:
        papers = papers + json.load(f) 
len(papers)
```

```python
authors_map = {}
for paper in papers:
    subject = paper["subject"]
    authors = paper["authors"]
    for author in authors:
        if author not in authors_map:
            authors_map[author] = {}
        if subject not in authors_map[author]:
            authors_map[author][subject] = 0
        authors_map[author][subject] += 1
len(authors_map)
```

```python
sorted(authors_map.items(), key=lambda x:len(x[1]), reverse=True)
```

```python
with open(f"data/authors_2000_to_2010.json", "w") as f:
    json.dump(authors_map, f, ensure_ascii=False, indent=4)
```
