```python
playlist_id = "37i9dQZEVXbMMy2roB9myp" # TOP 50 Argentina!
playlist_url = f"https://api.spotify.com/v1/playlists/{playlist_id}"
playlist = requests.get(playlist_url, headers=headers).json()
```

```python
artist_ids = []
for track in playlist["tracks"]["items"]:
    for artist in track["track"]["artists"]:
        if artist["id"] not in artist_ids:
            artist_ids.append(artist["id"])

artists_d0 = []
print("Processing depth 0: ", end="") # logging
for artist_id in artist_ids:
    print(artist_id, end=", ") # logging
    url = f"https://api.spotify.com/v1/artists/{artist_id}"
    artist = requests.get(url, headers=headers).json()
    artists_d0.append({
        "id": artist["id"],
        "name": artist["name"],
        "genres": artist["genres"],
        "popularity": artist["popularity"],
        "followers": artist["followers"]["total"],
        "related": []
    })
```

```python
artists_d1 = []
print("Processing depth 1: ", end="") # logging
for i in range(len(artists_d0)):
    artist_id = artists_d0[i]["id"]
    print(artist_id, end=", ") # logging
    url = f"https://api.spotify.com/v1/artists/{artist_id}/related-artists"
    related_artists = requests.get(url, headers=headers).json()["artists"]
    for related in related_artists:
        artists_d0[i]["related"].append(related["id"])
        if related["id"] not in artist_ids:
            artist_ids.append(related["id"])
            artists_d1.append({
                "id": related["id"],
                "name": related["name"],
                "genres": related["genres"],
                "popularity": related["popularity"],
                "followers": related["followers"]["total"],
                "related": []
             })
```

```python
artists_d2 = []
print("Processing depth 2: ", end="") # logging
for i in range(len(artists_d1)):
    artist_id = artists_d1[i]["id"]
    print(artist_id, end=", ") # logging
    url = f"https://api.spotify.com/v1/artists/{artist_id}/related-artists"
    related_artists = requests.get(url, headers=headers).json()["artists"]
    for related in related_artists:
        artists_d1[i]["related"].append(related["id"])
        if related["id"] not in artist_ids:
            artist_ids.append(related["id"])
            artists_d2.append({
                "id": related["id"],
                "name": related["name"],
                "genres": related["genres"],
                "popularity": related["popularity"],
                "followers": related["followers"]["total"],
                "related": []
             })
```

```python
artists_d3 = []
print("Processing depth 3: ", end="") # logging
for i in range(len(artists_d2)):
    artist_id = artists_d2[i]["id"]
    print(artist_id, end=", ") # logging
    url = f"https://api.spotify.com/v1/artists/{artist_id}/related-artists"
    related_artists = requests.get(url, headers=headers).json()["artists"]
    for related in related_artists:
        artists_d2[i]["related"].append(related["id"])
        if related["id"] not in artist_ids:
            artist_ids.append(related["id"])
            artists_d3.append({
                "id": related["id"],
                "name": related["name"],
                "genres": related["genres"],
                "popularity": related["popularity"],
                "followers": related["followers"]["total"],
                "related": []
             })
```

```python
d0, d1, d2 = [a["id"] for a in artists_d0], [a["id"] for a in artists_d1], [a["id"] for a in artists_d2]
artists = artists_d0+artists_d1+artists_d2
nodes_id = [a["id"] for a in artists]
with open("data/nodes.csv", "w") as nodes, open("data/edges.csv", "w") as edges:
    nodeswriter = csv.writer(nodes, delimiter=";")
    nodeswriter.writerow(["id", "name", "popularity", "followers", "d"])
    edgeswriter = csv.writer(edges, delimiter=';')
    edgeswriter.writerow(["Source", "Target"])

    for artist in artists:
        nodeswriter.writerow([
            artist["id"],
            artist["name"],
            artist["popularity"],
            artist["followers"],
            0 if (artist["id"] in d0) else (1 if (artist["id"] in d1) else 2)
        ])
        for related in artist["related"]:
            if related in nodes_id:
                edgeswriter.writerow([artist["id"], related])
```
